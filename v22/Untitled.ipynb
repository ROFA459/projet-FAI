```python
# ============================================
# üì¶ 1. ÿ™ÿ´ÿ®Ÿäÿ™ Ÿàÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÖŸÉÿ™ÿ®ÿßÿ™ ÿßŸÑŸÖÿ∑ŸÑŸàÿ®ÿ©
# ============================================
import subprocess
import sys

def install_packages():
    """ÿ™ÿ´ÿ®Ÿäÿ™ ÿßŸÑŸÖŸÉÿ™ÿ®ÿßÿ™ ÿßŸÑŸÖÿ∑ŸÑŸàÿ®ÿ© ÿ•ÿ∞ÿß ŸÑŸÖ ÿ™ŸÉŸÜ ŸÖÿ´ÿ®ÿ™ÿ©"""
    required_packages = [
        'torch', 'pandas', 'numpy', 'matplotlib',
        'seaborn', 'scikit-learn', 'scipy', 'joblib'
    ]
    
    for package in required_packages:
        try:
            if package == 'torch':
                __import__('torch')
            elif package == 'sklearn':
                __import__('sklearn')
            else:
                __import__(package)
            print(f"‚úÖ {package} already installed")
        except ImportError:
            print(f"üì¶ Installing {package}...")
            subprocess.check_call([sys.executable, "-m", "pip", "install", package, "--quiet"])

# ÿ™ÿ´ÿ®Ÿäÿ™ ÿßŸÑŸÖŸÉÿ™ÿ®ÿßÿ™
print("üîß Checking and installing required packages...")
install_packages()

# ============================================
# üìö 2. ÿßÿ≥ÿ™Ÿäÿ±ÿßÿØ ÿßŸÑŸÖŸÉÿ™ÿ®ÿßÿ™
# ============================================
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset, Dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import KFold
from scipy import stats
import joblib
import warnings
import time
import os
from datetime import datetime

warnings.filterwarnings('ignore')

# ÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑÿπÿ±ÿ∂
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.4f}'.format)
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("‚úÖ All libraries imported successfully!")

# ============================================
# üîç 3. ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ£ŸàŸÑÿßŸã
# ============================================
def analyze_data():
    """ÿ™ÿ≠ŸÑŸäŸÑ ÿ¥ÿßŸÖŸÑ ŸÑŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÇÿ®ŸÑ ÿßŸÑÿ®ÿØÿ°"""
    print("\n" + "="*60)
    print("üìä DATA ANALYSIS")
    print("="*60)
    
    # ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
    X_train = pd.read_csv("train_filtered.csv")
    y_train = pd.read_csv("y_train.csv")
    X_val = pd.read_csv("val_filtered.csv")
    y_val = pd.read_csv("y_val.csv")
    X_test = pd.read_csv("test_filtered.csv")
    y_test = pd.read_csv("y_test.csv")
    
    print(f"\nüìà Dataset Shapes:")
    print(f"  Training:   {X_train.shape}")
    print(f"  Validation: {X_val.shape}")
    print(f"  Test:       {X_test.shape}")
    
    print(f"\nüéØ Target Statistics (y_train):")
    print(f"  Mean:  {y_train.values.mean():.4f}")
    print(f"  Std:   {y_train.values.std():.4f}")
    print(f"  Min:   {y_train.values.min():.4f}")
    print(f"  Max:   {y_train.values.max():.4f}")
    print(f"  Skew:  {pd.Series(y_train.values.ravel()).skew():.4f}")
    
    print(f"\nüîç Features Analysis:")
    print(f"  Number of features: {X_train.shape[1]}")
    print(f"  Feature names: {list(X_train.columns)}")
    
    # ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸÖÿ™ÿ∑ÿ±ŸÅÿ©
    print(f"\n‚ö†Ô∏è  Outliers Detection (IQR method):")
    Q1 = X_train.quantile(0.25)
    Q3 = X_train.quantile(0.75)
    IQR = Q3 - Q1
    outliers_count = ((X_train < (Q1 - 1.5 * IQR)) | (X_train > (Q3 + 1.5 * IQR))).sum()
    print(outliers_count.sort_values(ascending=False).head())
    
    return X_train, y_train, X_val, y_val, X_test, y_test

# ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
X_train, y_train, X_val, y_val, X_test, y_test = analyze_data()

# ============================================
# ‚öôÔ∏è 4. ÿ•ÿπÿØÿßÿØ ÿßŸÑÿ¨Ÿáÿßÿ≤ Ÿàÿ™ŸáŸäÿ¶ÿ© ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ============================================
# ÿ™ÿ≠ÿØŸäÿØ ÿßŸÑÿ¨Ÿáÿßÿ≤ (GPU ÿ•ÿ∞ÿß ŸÖÿ™ŸàŸÅÿ±)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\nüíª Using device: {device}")
if torch.cuda.is_available():
    print(f"  GPU: {torch.cuda.get_device_name(0)}")
    print(f"  Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

# ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ•ŸÑŸâ tensors
def prepare_tensors(X_train, y_train, X_val, y_val, X_test, y_test):
    """ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ•ŸÑŸâ PyTorch tensors"""
    # ÿ™ÿ≠ŸàŸäŸÑ ÿ•ŸÑŸâ numpy arrays ÿ£ŸàŸÑÿßŸã
    X_train_np = X_train.values.astype(np.float32)
    y_train_np = y_train.values.astype(np.float32).reshape(-1, 1)
    X_val_np = X_val.values.astype(np.float32)
    y_val_np = y_val.values.astype(np.float32).reshape(-1, 1)
    X_test_np = X_test.values.astype(np.float32)
    y_test_np = y_test.values.astype(np.float32).reshape(-1, 1)
    
    # ÿ™ÿ≠ŸàŸäŸÑ ÿ•ŸÑŸâ PyTorch tensors
    X_train_tensor = torch.from_numpy(X_train_np).to(device)
    y_train_tensor = torch.from_numpy(y_train_np).to(device)
    X_val_tensor = torch.from_numpy(X_val_np).to(device)
    y_val_tensor = torch.from_numpy(y_val_np).to(device)
    X_test_tensor = torch.from_numpy(X_test_np).to(device)
    y_test_tensor = torch.from_numpy(y_test_np).to(device)
    
    return (X_train_tensor, y_train_tensor, X_val_tensor, y_val_tensor, 
            X_test_tensor, y_test_tensor)

# ÿ™ÿ≠ÿ∂Ÿäÿ± ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
tensors = prepare_tensors(X_train, y_train, X_val, y_val, X_test, y_test)
X_train_t, y_train_t, X_val_t, y_val_t, X_test_t, y_test_t = tensors

print(f"\n‚úÖ Data prepared successfully:")
print(f"  Training tensor shape:   {X_train_t.shape}")
print(f"  Validation tensor shape: {X_val_t.shape}")
print(f"  Test tensor shape:       {X_test_t.shape}")

# ============================================
# üß† 5. ÿ™ÿπÿ±ŸäŸÅ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨ ÿßŸÑÿπÿµÿ®Ÿäÿ© ÿßŸÑŸÖÿ™ŸÇÿØŸÖÿ© (ŸÖÿµÿ≠ÿ≠)
# ============================================
class AdvancedRealEstateNN(nn.Module):
    """ŸÜŸÖŸàÿ∞ÿ¨ ÿ¥ÿ®ŸÉÿ© ÿπÿµÿ®Ÿäÿ© ŸÖÿ™ŸÇÿØŸÖ ŸÑŸÑÿ™ŸÜÿ®ÿ§ ÿßŸÑÿπŸÇÿßÿ±Ÿä"""
    def __init__(self, input_size, hidden_layers=[128, 64, 32], dropout_rate=0.3):
        super(AdvancedRealEstateNN, self).__init__()
        
        layers = []
        prev_size = input_size
        
        # ÿ®ŸÜÿßÿ° ÿßŸÑÿ∑ÿ®ŸÇÿßÿ™ ÿßŸÑÿØŸäŸÜÿßŸÖŸäŸÉŸäÿ© (ŸÖÿµÿ∫ÿ± ŸÑŸÄ 5 ŸÖŸäÿ≤ÿßÿ™)
        for i, hidden_size in enumerate(hidden_layers):
            layers.append(nn.Linear(prev_size, hidden_size))
            layers.append(nn.BatchNorm1d(hidden_size))
            layers.append(nn.LeakyReLU(negative_slope=0.1))
            layers.append(nn.Dropout(dropout_rate))
            prev_size = hidden_size
        
        # ÿ∑ÿ®ŸÇÿ© ÿßŸÑÿ•ÿÆÿ±ÿßÿ¨
        layers.append(nn.Linear(prev_size, 1))
        
        self.network = nn.Sequential(*layers)
        
        # Ÿàÿ≤ŸÜ initialization
        self._initialize_weights()
    
    def _initialize_weights(self):
        """ÿ™ŸáŸäÿ¶ÿ© ÿßŸÑÿ£Ÿàÿ≤ÿßŸÜ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ Kaiming initialization"""
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_in', nonlinearity='leaky_relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm1d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)
    
    def forward(self, x):
        return self.network(x)

class ResidualRealEstateNN(nn.Module):
    """ŸÜŸÖŸàÿ∞ÿ¨ ŸÖÿπ ŸàÿµŸÑÿßÿ™ ŸÖÿ™ÿ®ŸÇŸäÿ© (ResNet-like) - ŸÖÿµŸÖŸÖ ŸÑŸÄ 5 ŸÖŸäÿ≤ÿßÿ™"""
    def __init__(self, input_size):
        super(ResidualRealEstateNN, self).__init__()
        
        # ÿ∑ÿ®ŸÇÿ© ÿßŸÑÿ•ÿØÿÆÿßŸÑ (5 ‚Üí 32)
        self.fc1 = nn.Linear(input_size, 32)
        self.bn1 = nn.BatchNorm1d(32)
        
        # ÿ∑ÿ®ŸÇÿ© ŸÖÿÆŸÅŸäÿ© (32 ‚Üí 16)
        self.fc2 = nn.Linear(32, 16)
        self.bn2 = nn.BatchNorm1d(16)
        
        # ÿ∑ÿ®ŸÇÿ© ŸÖÿπ residual connection (16 ‚Üí 16)
        self.fc3 = nn.Linear(16, 16)
        self.bn3 = nn.BatchNorm1d(16)
        
        # ÿ∑ÿ®ŸÇÿ© ÿßŸÑÿ•ÿÆÿ±ÿßÿ¨
        self.fc4 = nn.Linear(16, 1)
        
        self.dropout = nn.Dropout(0.2)
        self.activation = nn.LeakyReLU(0.1)
        
        # ÿ∑ÿ®ŸÇÿ© projection ŸÑŸÑŸÄ residual ÿ•ÿ∞ÿß ŸÉÿßŸÜÿ™ ÿßŸÑÿ£ÿ®ÿπÿßÿØ ŸÖÿÆÿ™ŸÑŸÅÿ©
        self.residual_projection = nn.Linear(32, 16) if input_size != 16 else nn.Identity()
        
        # ÿ≠ŸÅÿ∏ input_size ŸÑŸÑÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ŸÅŸä forward
        self.input_size = input_size
    
    def forward(self, x):
        # ÿ∑ÿ®ŸÇÿ© ÿßŸÑÿ•ÿØÿÆÿßŸÑ
        x1 = self.activation(self.bn1(self.fc1(x)))
        x1 = self.dropout(x1)
        
        # ÿ∑ÿ®ŸÇÿ© ŸÖÿÆŸÅŸäÿ©
        x2 = self.activation(self.bn2(self.fc2(x1)))
        x2 = self.dropout(x2)
        
        # Residual connection
        # ŸÜÿ≠ÿ™ÿßÿ¨ ŸÑÿ™ÿ≠ŸàŸäŸÑ x1 (32) ÿ•ŸÑŸâ 16 ŸÑŸäÿ™ŸÖŸÉŸÜ ŸÖŸÜ ÿßŸÑÿ¨ŸÖÿπ ŸÖÿπ x2 (16)
        if self.input_size != 16:
            residual = self.residual_projection(x1)
        else:
            residual = x1[:, :16] if x1.shape[1] > 16 else x1
        
        # ÿ™ÿ∑ÿßÿ®ŸÇ ÿßŸÑÿ£ÿ®ÿπÿßÿØ
        if residual.shape[1] != x2.shape[1]:
            # ÿ•ÿ∞ÿß ŸÉÿßŸÜÿ™ ÿßŸÑÿ£ÿ®ÿπÿßÿØ ŸÑÿß ÿ™ÿ≤ÿßŸÑ ÿ∫Ÿäÿ± ŸÖÿ™ÿ∑ÿßÿ®ŸÇÿ©ÿå ŸÜŸÇŸàŸÖ ÿ®ÿ™ÿπÿØŸäŸÑ
            if residual.shape[1] < x2.shape[1]:
                # padding
                padding = torch.zeros(residual.shape[0], x2.shape[1] - residual.shape[1], 
                                    device=residual.device)
                residual = torch.cat([residual, padding], dim=1)
            else:
                # trimming
                residual = residual[:, :x2.shape[1]]
        
        # ÿ¨ŸÖÿπ ŸÖÿπ residual
        x3 = self.activation(self.bn3(self.fc3(x2) + residual))
        x3 = self.dropout(x3)
        
        # ÿßŸÑÿ•ÿÆÿ±ÿßÿ¨
        output = self.fc4(x3)
        
        return output

class SimpleRealEstateNN(nn.Module):
    """ŸÜŸÖŸàÿ∞ÿ¨ ÿ®ÿ≥Ÿäÿ∑ ŸÑŸÉŸÜ ŸÅÿπÿßŸÑ ŸÑÿπÿØÿØ ŸÇŸÑŸäŸÑ ŸÖŸÜ ÿßŸÑŸÖŸäÿ≤ÿßÿ™"""
    def __init__(self, input_size):
        super(SimpleRealEstateNN, self).__init__()
        
        self.network = nn.Sequential(
            nn.Linear(input_size, 64),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.2),
            
            nn.Linear(64, 32),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.2),
            
            nn.Linear(32, 16),
            nn.LeakyReLU(0.1),
            
            nn.Linear(16, 1)
        )
    
    def forward(self, x):
        return self.network(x)

# ============================================
# üèãÔ∏è 6. ŸÅÿ¶ÿ© ÿßŸÑÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÖÿ™ŸÇÿØŸÖÿ©
# ============================================
class AdvancedTrainer:
    """ŸÅÿ¶ÿ© ŸÖÿ™ŸÇÿØŸÖÿ© ŸÑŸÑÿ™ÿØÿ±Ÿäÿ® ŸÖÿπ ŸÖŸäÿ≤ÿßÿ™ ŸÖÿ™ÿπÿØÿØÿ©"""
    def __init__(self, model, model_name="RealEstateModel"):
        self.model = model.to(device)
        self.model_name = model_name
        self.train_losses = []
        self.val_losses = []
        self.learning_rates = []
        self.best_val_loss = float('inf')
        self.best_model_state = None
        self.start_time = None
        
    def train_epoch(self, train_loader, criterion, optimizer, scheduler=None):
        """ÿ™ÿØÿ±Ÿäÿ® ŸÑŸÖÿØÿ© epoch Ÿàÿßÿ≠ÿØÿ©"""
        self.model.train()
        running_loss = 0.0
        
        for batch_idx, (data, target) in enumerate(train_loader):
            optimizer.zero_grad()
            
            # Forward pass
            output = self.model(data)
            loss = criterion(output, target)
            
            # ÿ•ÿ∂ÿßŸÅÿ© L1 regularization
            l1_lambda = 0.001
            l1_norm = sum(p.abs().sum() for p in self.model.parameters())
            loss = loss + l1_lambda * l1_norm
            
            # Backward pass
            loss.backward()
            
            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
            
            optimizer.step()
            running_loss += loss.item()
            
            if scheduler:
                scheduler.step()
        
        return running_loss / len(train_loader)
    
    def validate(self, val_loader, criterion):
        """ÿßŸÑÿ™ÿ≠ŸÇŸÇ ŸÖŸÜ ÿ£ÿØÿßÿ° ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨"""
        self.model.eval()
        val_loss = 0.0
        
        with torch.no_grad():
            for data, target in val_loader:
                output = self.model(data)
                val_loss += criterion(output, target).item()
        
        return val_loss / len(val_loader)
    
    def train(self, train_loader, val_loader, epochs=300, 
              learning_rate=0.001, weight_decay=0.0001,
              patience=30, min_delta=0.001):
        """ÿØŸàÿ±ÿ© ÿßŸÑÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÉÿßŸÖŸÑÿ©"""
        print(f"\nüöÄ Starting training for {self.model_name}...")
        print(f"  Epochs: {epochs}, Learning rate: {learning_rate}")
        print("="*60)
        
        self.start_time = time.time()
        
        # Loss function
        criterion = nn.HuberLoss(delta=1.0)  # ÿ£ŸÉÿ´ÿ± ŸÇŸàÿ© ŸÖŸÜ MSE
        
        # Optimizer ŸÖÿπ weight decay
        optimizer = optim.AdamW(self.model.parameters(), lr=learning_rate, 
                               weight_decay=weight_decay)
        
        # Learning rate scheduler
        scheduler = optim.lr_scheduler.OneCycleLR(
            optimizer, 
            max_lr=learning_rate,
            epochs=epochs,
            steps_per_epoch=len(train_loader),
            pct_start=0.3
        )
        
        # Early stopping
        early_stop_counter = 0
        
        for epoch in range(epochs):
            # ÿßŸÑÿ™ÿØÿ±Ÿäÿ®
            train_loss = self.train_epoch(train_loader, criterion, optimizer, scheduler)
            self.train_losses.append(train_loss)
            
            # ÿßŸÑÿ™ÿ≠ŸÇŸÇ
            val_loss = self.validate(val_loader, criterion)
            self.val_losses.append(val_loss)
            
            # ÿ≠ŸÅÿ∏ ŸÖÿπÿØŸÑ ÿßŸÑÿ™ÿπŸÑŸÖ
            self.learning_rates.append(optimizer.param_groups[0]['lr'])
            
            # ÿ≠ŸÅÿ∏ ÿ£ŸÅÿ∂ŸÑ ŸÜŸÖŸàÿ∞ÿ¨
            if val_loss < self.best_val_loss - min_delta:
                self.best_val_loss = val_loss
                self.best_model_state = self.model.state_dict().copy()
                early_stop_counter = 0
                
                # ÿ≠ŸÅÿ∏ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
                torch.save({
                    'epoch': epoch,
                    'model_state_dict': self.best_model_state,
                    'optimizer_state_dict': optimizer.state_dict(),
                    'val_loss': val_loss,
                    'train_loss': train_loss,
                }, f'best_{self.model_name}.pth')
            else:
                early_stop_counter += 1
            
            # ÿ∑ÿ®ÿßÿπÿ© ÿßŸÑÿ™ŸÇÿØŸÖ
            if (epoch + 1) % 20 == 0 or epoch == 0 or epoch == epochs - 1:
                elapsed = time.time() - self.start_time
                eta = (epochs - epoch - 1) * (elapsed / (epoch + 1))
                print(f"Epoch [{epoch+1:03d}/{epochs}] | "
                      f"Train Loss: {train_loss:.6f} | "
                      f"Val Loss: {val_loss:.6f} | "
                      f"LR: {self.learning_rates[-1]:.6f} | "
                      f"ETA: {eta/60:.1f}m")
            
            # Early stopping
            if early_stop_counter >= patience:
                print(f"\n‚ö†Ô∏è  Early stopping triggered at epoch {epoch+1}")
                break
        
        # ÿ™ÿ≠ŸÖŸäŸÑ ÿ£ŸÅÿ∂ŸÑ ŸÜŸÖŸàÿ∞ÿ¨
        if self.best_model_state:
            self.model.load_state_dict(self.best_model_state)
        
        training_time = time.time() - self.start_time
        print(f"\n‚úÖ Training completed in {training_time/60:.2f} minutes")
        print(f"üìâ Best validation loss: {self.best_val_loss:.6f}")
        
        return self.model
    
    def plot_training_history(self):
        """ÿ±ÿ≥ŸÖ ŸÖŸÜÿ≠ŸÜŸâ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
        
        # ÿ±ÿ≥ŸÖ ÿßŸÑÿÆÿ≥ÿßÿ¶ÿ±
        ax1.plot(self.train_losses, label='Training Loss', linewidth=2)
        ax1.plot(self.val_losses, label='Validation Loss', linewidth=2)
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.set_title(f'{self.model_name} - Training History')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # ÿ±ÿ≥ŸÖ ŸÖÿπÿØŸÑ ÿßŸÑÿ™ÿπŸÑŸÖ
        ax2.plot(self.learning_rates, color='green', linewidth=2)
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Learning Rate')
        ax2.set_title('Learning Rate Schedule')
        ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(f'training_history_{self.model_name}.png', dpi=120, bbox_inches='tight')
        plt.show()

# ============================================
# üîÑ 7. ÿ•ŸÜÿ¥ÿßÿ° DataLoaders
# ============================================
class RealEstateDataset(Dataset):
    """Dataset ŸÖÿÆÿµÿµ ŸÑŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿπŸÇÿßÿ±Ÿäÿ©"""
    def __init__(self, features, targets):
        self.features = features
        self.targets = targets
    
    def __len__(self):
        return len(self.features)
    
    def __getitem__(self, idx):
        return self.features[idx], self.targets[idx]

# ÿ•ŸÜÿ¥ÿßÿ° datasets
train_dataset = RealEstateDataset(X_train_t, y_train_t)
val_dataset = RealEstateDataset(X_val_t, y_val_t)
test_dataset = RealEstateDataset(X_test_t, y_test_t)

# ÿ•ŸÜÿ¥ÿßÿ° data loaders
batch_size = min(64, len(train_dataset) // 10)  # ÿØŸäŸÜÿßŸÖŸäŸÉŸä ÿ≠ÿ≥ÿ® ÿ≠ÿ¨ŸÖ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=0)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=0)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=0)

print(f"\nüìä DataLoaders created:")
print(f"  Batch size: {batch_size}")
print(f"  Training batches: {len(train_loader)}")
print(f"  Validation batches: {len(val_loader)}")
print(f"  Test batches: {len(test_loader)}")

# ============================================
# ü§ñ 8. ÿ™ÿØÿ±Ÿäÿ® ŸÜŸÖÿßÿ∞ÿ¨ ŸÖÿ™ÿπÿØÿØÿ© (ŸÖÿµÿ≠ÿ≠)
# ============================================
print("\n" + "="*60)
print("üß† TRAINING MULTIPLE MODELS")
print("="*60)

# ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨ ÿßŸÑÿ™Ÿä ÿ≥ŸÜÿØÿ±ÿ®Ÿáÿß (ŸÖÿµŸÖŸÖÿ© ÿÆÿµŸäÿµÿßŸã ŸÑŸÄ 5 ŸÖŸäÿ≤ÿßÿ™)
input_size = X_train.shape[1]
print(f"‚ö†Ô∏è  Note: Only {input_size} features available")

models_config = [
    {
        'model': AdvancedRealEstateNN(input_size, hidden_layers=[32, 16, 8], dropout_rate=0.2),
        'name': 'AdvancedModel_Small',
        'epochs': 200,
        'lr': 0.001
    },
    {
        'model': SimpleRealEstateNN(input_size),
        'name': 'SimpleModel',
        'epochs': 150,
        'lr': 0.0015
    },
    {
        'model': ResidualRealEstateNN(input_size),
        'name': 'ResidualModel',
        'epochs': 250,
        'lr': 0.0008
    }
]

# ÿ™ÿØÿ±Ÿäÿ® ÿ¨ŸÖŸäÿπ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
trained_models = {}
for config in models_config:
    print(f"\nüî• Training {config['name']}...")
    print(f"  Architecture: {config['model']}")
    
    try:
        # ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÖÿØÿ±ÿ®
        trainer = AdvancedTrainer(config['model'], config['name'])
        
        # ÿßŸÑÿ™ÿØÿ±Ÿäÿ®
        model = trainer.train(
            train_loader, 
            val_loader, 
            epochs=config['epochs'],
            learning_rate=config['lr']
        )
        
        # ÿ±ÿ≥ŸÖ ÿ™ÿßÿ±ŸäÿÆ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®
        trainer.plot_training_history()
        
        # ÿ≠ŸÅÿ∏ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿßŸÑŸÖÿØÿ±ÿ®
        trained_models[config['name']] = {
            'model': model,
            'trainer': trainer
        }
        
        print(f"‚úÖ {config['name']} trained successfully!")
        
    except Exception as e:
        print(f"‚ùå Error training {config['name']}: {e}")
        print("Skipping to next model...")

# ============================================
# üìä 9. ÿ™ŸÇŸäŸäŸÖ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨ (ŸÖÿµÿ≠ÿ≠)
# ============================================
print("\n" + "="*60)
print("üìà EVALUATING ALL MODELS")
print("="*60)

def evaluate_model(model, test_loader, model_is_function=False):
    """ÿ™ŸÇŸäŸäŸÖ ÿ¥ÿßŸÖŸÑ ŸÑŸÑŸÜŸÖŸàÿ∞ÿ¨"""
    if not model_is_function:
        model.eval()
    
    all_predictions = []
    all_targets = []
    
    with torch.no_grad():
        for data, target in test_loader:
            if model_is_function:
                # ÿ•ÿ∞ÿß ŸÉÿßŸÜ model ÿØÿßŸÑÿ© lambda
                output = model(data)
            else:
                # ÿ•ÿ∞ÿß ŸÉÿßŸÜ model ŸÉÿßÿ¶ŸÜ nn.Module
                output = model(data)
            
            all_predictions.extend(output.cpu().numpy())
            all_targets.extend(target.cpu().numpy())
    
    all_predictions = np.array(all_predictions).flatten()
    all_targets = np.array(all_targets).flatten()
    
    # ÿ≠ÿ≥ÿßÿ® ÿßŸÑŸÖŸÇÿßŸäŸäÿ≥
    metrics = {
        'MSE': mean_squared_error(all_targets, all_predictions),
        'RMSE': np.sqrt(mean_squared_error(all_targets, all_predictions)),
        'MAE': mean_absolute_error(all_targets, all_predictions),
        'R2': r2_score(all_targets, all_predictions),
        'MAPE': np.mean(np.abs((all_targets - all_predictions) / np.clip(np.abs(all_targets), 1e-8, None))) * 100
    }
    
    return metrics, all_predictions, all_targets

# ÿ™ŸÇŸäŸäŸÖ ÿ¨ŸÖŸäÿπ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
results = {}
all_model_predictions = {}

for name, model_info in trained_models.items():
    print(f"\nüìä Evaluating {name}...")
    
    metrics, predictions, targets = evaluate_model(model_info['model'], test_loader)
    results[name] = metrics
    all_model_predictions[name] = predictions
    
    print(f"  R¬≤ Score:   {metrics['R2']:.6f}")
    print(f"  RMSE:       {metrics['RMSE']:.6f}")
    print(f"  MAE:        {metrics['MAE']:.6f}")
    print(f"  MAPE:       {metrics['MAPE']:.2f}%")

# ============================================
# ü§ù 10. Ensemble Model (ŸÖÿµÿ≠ÿ≠)
# ============================================
print("\n" + "="*60)
print("üèÜ ENSEMBLE MODEL (Weighted Average)")
print("="*60)

# ÿ¨ŸÖÿπ ÿ™ŸÜÿ®ÿ§ÿßÿ™ ÿ¨ŸÖŸäÿπ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
ensemble_predictions = np.zeros_like(all_model_predictions[list(trained_models.keys())[0]])

# ÿ£Ÿàÿ≤ÿßŸÜ ÿ™ÿπÿ™ŸÖÿØ ÿπŸÑŸâ ÿ£ÿØÿßÿ° R¬≤
weights = {}
total_r2 = sum(results[name]['R2'] for name in results)
for name in results:
    weights[name] = results[name]['R2'] / total_r2
    ensemble_predictions += all_model_predictions[name] * weights[name]

# ÿ™ŸÇŸäŸäŸÖ Ensemble - ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿØÿßŸÑÿ© ÿ™ŸÇŸäŸäŸÖ ŸÖÿÆÿµÿµÿ© ŸÑŸÑŸÄ Ensemble
def evaluate_ensemble_predictions(predictions, test_loader):
    """ÿ™ŸÇŸäŸäŸÖ ÿ™ŸÜÿ®ÿ§ÿßÿ™ Ensemble"""
    all_targets = []
    
    with torch.no_grad():
        for _, target in test_loader:
            all_targets.extend(target.cpu().numpy())
    
    all_targets = np.array(all_targets).flatten()
    
    # ÿ≠ÿ≥ÿßÿ® ÿßŸÑŸÖŸÇÿßŸäŸäÿ≥
    metrics = {
        'MSE': mean_squared_error(all_targets, predictions),
        'RMSE': np.sqrt(mean_squared_error(all_targets, predictions)),
        'MAE': mean_absolute_error(all_targets, predictions),
        'R2': r2_score(all_targets, predictions),
        'MAPE': np.mean(np.abs((all_targets - predictions) / np.clip(np.abs(all_targets), 1e-8, None))) * 100
    }
    
    return metrics

# ÿßŸÑÿ≠ÿµŸàŸÑ ÿπŸÑŸâ y_test ŸÖŸÜ test_loader
y_test_np = []
with torch.no_grad():
    for _, target in test_loader:
        y_test_np.extend(target.cpu().numpy())
y_test_np = np.array(y_test_np).flatten()

# ÿ™ŸÇŸäŸäŸÖ Ensemble
ensemble_metrics = evaluate_ensemble_predictions(ensemble_predictions, test_loader)
results['Ensemble'] = ensemble_metrics

print(f"\nüéØ Ensemble Model Results:")
print(f"  R¬≤ Score:   {ensemble_metrics['R2']:.6f}")
print(f"  RMSE:       {ensemble_metrics['RMSE']:.6f}")
print(f"  MAE:        {ensemble_metrics['MAE']:.6f}")
print(f"  MAPE:       {ensemble_metrics['MAPE']:.2f}%")

print(f"\nüìä Model Weights:")
for name, weight in weights.items():
    print(f"  {name}: {weight:.3f} (R¬≤: {results[name]['R2']:.4f})")

# ============================================
# üìà 11. ÿπÿ±ÿ∂ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨ ÿ®ŸäÿßŸÜŸäÿßŸã
# ============================================
print("\n" + "="*60)
print("üé® VISUALIZING RESULTS")
print("="*60)

def create_comprehensive_visualizations(results, all_model_predictions, y_test_np):
    """ÿ•ŸÜÿ¥ÿßÿ° ÿ±ÿ≥ŸàŸÖ ÿ®ŸäÿßŸÜŸäÿ© ÿ¥ÿßŸÖŸÑÿ©"""
    fig = plt.figure(figsize=(20, 15))
    
    # 1. ŸÖŸÇÿßÿ±ŸÜÿ© ÿ£ÿØÿßÿ° ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
    ax1 = plt.subplot(3, 3, 1)
    model_names = list(results.keys())
    r2_scores = [results[name]['R2'] for name in model_names]
    rmse_scores = [results[name]['RMSE'] for name in model_names]
    
    x = np.arange(len(model_names))
    width = 0.35
    
    bars1 = ax1.bar(x - width/2, r2_scores, width, label='R¬≤ Score', color='skyblue')
    bars2 = ax1.bar(x + width/2, rmse_scores, width, label='RMSE', color='lightcoral')
    
    ax1.set_xlabel('Models')
    ax1.set_title('Model Performance Comparison')
    ax1.set_xticks(x)
    ax1.set_xticklabels(model_names, rotation=45, ha='right')
    ax1.legend()
    ax1.grid(True, alpha=0.3, axis='y')
    
    # ÿ•ÿ∂ÿßŸÅÿ© ÿßŸÑŸÇŸäŸÖ ÿπŸÑŸâ ÿßŸÑÿ£ÿπŸÖÿØÿ©
    for bars in [bars1, bars2]:
        for bar in bars:
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:.3f}', ha='center', va='bottom', fontsize=8)
    
    # 2. ÿßŸÑÿ™ŸÜÿ®ÿ§ÿßÿ™ ŸÖŸÇÿßÿ®ŸÑ ÿßŸÑŸÇŸäŸÖ ÿßŸÑÿ≠ŸÇŸäŸÇŸäÿ© (Ensemble)
    ax2 = plt.subplot(3, 3, 2)
    y_pred_ensemble = all_model_predictions.get('Ensemble', ensemble_predictions)
    ax2.scatter(y_test_np, y_pred_ensemble, alpha=0.6, s=30)
    
    min_val = min(y_test_np.min(), y_pred_ensemble.min())
    max_val = max(y_test_np.max(), y_pred_ensemble.max())
    ax2.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2)
    
    ax2.set_xlabel('Actual Values')
    ax2.set_ylabel('Predicted Values')
    ax2.set_title(f'Ensemble Predictions\nR¬≤ = {results.get("Ensemble", {}).get("R2", 0):.4f}')
    ax2.grid(True, alpha=0.3)
    
    # 3. ÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑÿ£ÿÆÿ∑ÿßÿ°
    ax3 = plt.subplot(3, 3, 3)
    errors = y_pred_ensemble - y_test_np
    ax3.hist(errors, bins=30, edgecolor='black', alpha=0.7, density=True)
    
    # ÿ•ÿ∂ÿßŸÅÿ© ŸÖŸÜÿ≠ŸÜŸâ ÿßŸÑÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑÿ∑ÿ®ŸäÿπŸä
    mu, sigma = errors.mean(), errors.std()
    x_normal = np.linspace(errors.min(), errors.max(), 100)
    y_normal = (1/(sigma * np.sqrt(2*np.pi))) * np.exp(-0.5*((x_normal-mu)/sigma)**2)
    ax3.plot(x_normal, y_normal, 'r-', linewidth=2)
    
    ax3.axvline(x=0, color='g', linestyle='--', linewidth=2)
    ax3.set_xlabel('Prediction Error')
    ax3.set_ylabel('Density')
    ax3.set_title(f'Error Distribution\nMean: {mu:.3f}, Std: {sigma:.3f}')
    ax3.grid(True, alpha=0.3)
    
    # 4. Residual Plot
    ax4 = plt.subplot(3, 3, 4)
    ax4.scatter(y_pred_ensemble, errors, alpha=0.6, s=30)
    ax4.axhline(y=0, color='r', linestyle='--', linewidth=2)
    ax4.set_xlabel('Predicted Values')
    ax4.set_ylabel('Residuals')
    ax4.set_title('Residual Plot')
    ax4.grid(True, alpha=0.3)
    
    # 5. ŸÖŸÇÿßÿ±ŸÜÿ© ÿßŸÑÿ™ŸÜÿ®ÿ§ÿßÿ™ ŸÑŸÑŸÜŸÖÿßÿ∞ÿ¨ ÿßŸÑŸÖÿÆÿ™ŸÑŸÅÿ©
    ax5 = plt.subplot(3, 3, 5)
    sample_indices = np.arange(min(20, len(y_test_np)))
    
    for i, (name, preds) in enumerate(all_model_predictions.items()):
        ax5.plot(sample_indices, preds[sample_indices], 'o-', label=name, alpha=0.7)
    
    ax5.plot(sample_indices, y_test_np[sample_indices], 'ks-', label='Actual', linewidth=2, markersize=8)
    ax5.set_xlabel('Sample Index')
    ax5.set_ylabel('Value')
    ax5.set_title('Model Predictions Comparison')
    ax5.legend(loc='upper right', fontsize=8)
    ax5.grid(True, alpha=0.3)
    
    # 6. ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ÿßŸÑŸÖŸáŸÖÿ© (ÿßŸÅÿ™ÿ±ÿßÿ∂Ÿä)
    ax6 = plt.subplot(3, 3, 6)
    if hasattr(trained_models[list(trained_models.keys())[0]]['model'], 'network'):
        # ŸÖÿ≠ÿßŸàŸÑÿ© ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿ£Ÿàÿ≤ÿßŸÜ ÿßŸÑÿ∑ÿ®ŸÇÿ© ÿßŸÑÿ£ŸàŸÑŸâ
        try:
            first_layer_weights = trained_models[list(trained_models.keys())[0]]['model'].network[0].weight.data.cpu().numpy()
            feature_importance = np.abs(first_layer_weights).mean(axis=0)
            
            sorted_idx = np.argsort(feature_importance)[-10:]  # ÿ£ŸáŸÖ 10 ŸÖŸäÿ≤ÿßÿ™
            ax6.barh(range(len(sorted_idx)), feature_importance[sorted_idx])
            ax6.set_yticks(range(len(sorted_idx)))
            ax6.set_yticklabels([X_train.columns[i] for i in sorted_idx], fontsize=8)
            ax6.set_xlabel('Average Absolute Weight')
            ax6.set_title('Feature Importance (First Layer)')
        except:
            ax6.text(0.5, 0.5, 'Feature importance\nnot available', 
                    ha='center', va='center', transform=ax6.transAxes)
    else:
        ax6.text(0.5, 0.5, 'Feature importance\nnot available', 
                ha='center', va='center', transform=ax6.transAxes)
    
    ax6.grid(True, alpha=0.3, axis='x')
    
    # 7. Box plot ŸÑŸÑÿ£ÿÆÿ∑ÿßÿ°
    ax7 = plt.subplot(3, 3, 7)
    error_data = []
    labels = []
    for name, preds in all_model_predictions.items():
        error_data.append(preds - y_test_np)
        labels.append(name)
    
    ax7.boxplot(error_data, labels=labels)
    ax7.axhline(y=0, color='r', linestyle='--', linewidth=2)
    ax7.set_ylabel('Prediction Error')
    ax7.set_title('Error Distribution by Model')
    ax7.tick_params(axis='x', rotation=45)
    ax7.grid(True, alpha=0.3, axis='y')
    
    # 8. ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿ£ÿØÿßÿ° ŸÖŸÇÿßÿ®ŸÑ ÿ≠ÿ¨ŸÖ ÿßŸÑÿ™ŸÜÿ®ÿ§
    ax8 = plt.subplot(3, 3, 8)
    ax8.scatter(y_test_np, np.abs(errors), alpha=0.6, s=30)
    ax8.set_xlabel('Actual Value')
    ax8.set_ylabel('Absolute Error')
    ax8.set_title('Error vs Actual Value')
    ax8.grid(True, alpha=0.3)
    
    # 9. Summary text
    ax9 = plt.subplot(3, 3, 9)
    ax9.axis('off')
    
    summary_text = f"""
    MODEL PERFORMANCE SUMMARY
    {'='*30}
    
    Best Model: {max(results.items(), key=lambda x: x[1]['R2'])[0]}
    Best R¬≤: {max(r['R2'] for r in results.values()):.4f}
    
    Ensemble Results:
    ‚Ä¢ R¬≤: {results.get('Ensemble', {}).get('R2', 0):.4f}
    ‚Ä¢ RMSE: {results.get('Ensemble', {}).get('RMSE', 0):.4f}
    ‚Ä¢ MAE: {results.get('Ensemble', {}).get('MAE', 0):.4f}
    
    Dataset Info:
    ‚Ä¢ Features: {input_size}
    ‚Ä¢ Train samples: {len(X_train)}
    ‚Ä¢ Test samples: {len(X_test)}
    ‚Ä¢ Device: {device}
    """
    
    ax9.text(0.05, 0.95, summary_text, transform=ax9.transAxes,
            fontsize=10, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.suptitle('Real Estate Neural Network - Comprehensive Analysis', 
                fontsize=16, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.savefig('comprehensive_results.png', dpi=150, bbox_inches='tight')
    plt.show()

# ÿ™ÿ≠ŸàŸäŸÑ y_test ÿ•ŸÑŸâ numpy
y_test_np = y_test.values.ravel()

# ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑÿ±ÿ≥ŸàŸÖ ÿßŸÑÿ®ŸäÿßŸÜŸäÿ©
create_comprehensive_visualizations(results, all_model_predictions, y_test_np)

# ============================================
# üíæ 12. ÿ≠ŸÅÿ∏ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨ ŸàÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
# ============================================
print("\n" + "="*60)
print("üíæ SAVING MODELS AND RESULTS")
print("="*60)

# 1. ÿ≠ŸÅÿ∏ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿßŸÑŸÜŸáÿßÿ¶Ÿä (Ensemble)
ensemble_model_state = {
    'model_states': {},
    'weights': weights,
    'input_size': input_size,
    'feature_names': list(X_train.columns),
    'results': results,
    'ensemble_predictions': ensemble_predictions,
    'test_targets': y_test_np,
    'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
}

# ÿ≠ŸÅÿ∏ ÿ≠ÿßŸÑÿßÿ™ ÿ¨ŸÖŸäÿπ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
for name, model_info in trained_models.items():
    ensemble_model_state['model_states'][name] = model_info['model'].state_dict()

torch.save(ensemble_model_state, 'real_estate_ensemble_final.pth')

# 2. ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ŸÇŸäŸäŸÖ
results_df = pd.DataFrame(results).T
results_df.to_csv('model_performance_results.csv')
print("‚úÖ Model performance results saved to 'model_performance_results.csv'")

# 3. ÿ≠ŸÅÿ∏ ÿßŸÑÿ™ŸÜÿ®ÿ§ÿßÿ™
predictions_df = pd.DataFrame({
    'Actual': y_test_np,
    'Ensemble_Prediction': ensemble_predictions
})

# ÿ•ÿ∂ÿßŸÅÿ© ÿ™ŸÜÿ®ÿ§ÿßÿ™ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨ ÿßŸÑŸÅÿ±ÿØŸäÿ©
for name, preds in all_model_predictions.items():
    if name != 'Ensemble':  # Ensemble ŸÖŸàÿ¨ŸàÿØ ÿ®ÿßŸÑŸÅÿπŸÑ
        predictions_df[f'{name}_Prediction'] = preds

predictions_df.to_csv('all_predictions.csv', index=False)
print("‚úÖ All predictions saved to 'all_predictions.csv'")

# 4. ÿ•ŸÜÿ¥ÿßÿ° ÿ™ŸÇÿ±Ÿäÿ± ŸÜŸáÿßÿ¶Ÿä
report_content = f"""
{'='*60}
REAL ESTATE NEURAL NETWORK - FINAL REPORT
{'='*60}

üìÖ Generated on: {datetime.now().strftime("%Y-%m-%d %H:%M:%S")}
üíª Device used: {device}

üìä DATASET INFORMATION:
‚Ä¢ Total features: {input_size}
‚Ä¢ Training samples: {len(X_train)}
‚Ä¢ Validation samples: {len(X_val)}
‚Ä¢ Test samples: {len(X_test)}
‚Ä¢ Feature names: {list(X_train.columns)}

üèÜ BEST MODEL PERFORMANCE:
{results_df.loc[results_df['R2'].idxmax()].to_string()}

ü§ù ENSEMBLE MODEL DETAILS:
‚Ä¢ R¬≤ Score: {results.get('Ensemble', {}).get('R2', 0):.4f}
‚Ä¢ RMSE: {results.get('Ensemble', {}).get('RMSE', 0):.4f}
‚Ä¢ MAE: {results.get('Ensemble', {}).get('MAE', 0):.4f}
‚Ä¢ MAPE: {results.get('Ensemble', {}).get('MAPE', 0):.2f}%

üìà PERFORMANCE COMPARISON:
{results_df.to_string()}

üíæ SAVED FILES:
1. real_estate_ensemble_final.pth - Final ensemble model
2. model_performance_results.csv - Performance metrics
3. all_predictions.csv - All model predictions
4. comprehensive_results.png - Visualization
5. training_history_*.png - Training histories

üéØ RECOMMENDATIONS:
"""

# ÿ•ÿ∂ÿßŸÅÿ© ÿ™ŸàÿµŸäÿßÿ™ ÿ®ŸÜÿßÿ°Ÿã ÿπŸÑŸâ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
best_r2 = results_df['R2'].max()
if best_r2 > 0.8:
    report_content += "‚úÖ Excellent model performance! Ready for production use.\n"
elif best_r2 > 0.7:
    report_content += "‚ö†Ô∏è  Good performance. Consider feature engineering for further improvement.\n"
elif best_r2 > 0.6:
    report_content += "‚ö†Ô∏è  Fair performance. Try collecting more data or using different architectures.\n"
else:
    report_content += "‚ùå Needs improvement. Review data quality and preprocessing steps.\n"

report_content += "\n" + "="*60

# ÿ≠ŸÅÿ∏ ÿßŸÑÿ™ŸÇÿ±Ÿäÿ±
with open('final_report.txt', 'w', encoding='utf-8') as f:
    f.write(report_content)

print("‚úÖ Final report saved to 'final_report.txt'")

# ============================================
# üîÆ 13. ÿØÿßŸÑÿ© ŸÑŸÑÿ™ŸÜÿ®ÿ§ ÿπŸÑŸâ ÿ®ŸäÿßŸÜÿßÿ™ ÿ¨ÿØŸäÿØÿ©
# ============================================
class RealEstatePredictor:
    """ŸÅÿ¶ÿ© ŸÑŸÑÿ™ŸÜÿ®ÿ§ ÿπŸÑŸâ ÿ®ŸäÿßŸÜÿßÿ™ ÿ¨ÿØŸäÿØÿ©"""
    def __init__(self, model_path='real_estate_ensemble_final.pth'):
        # ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿßŸÑŸÖÿ≠ŸÅŸàÿ∏
        checkpoint = torch.load(model_path, map_location=device)
        
        self.input_size = checkpoint['input_size']
        self.feature_names = checkpoint['feature_names']
        self.weights = checkpoint['weights']
        self.results = checkpoint['results']
        
        # ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
        self.models = {}
        for name, state_dict in checkpoint['model_states'].items():
            if 'Advanced' in name:
                model = AdvancedRealEstateNN(self.input_size)
            elif 'Residual' in name:
                model = ResidualRealEstateNN(self.input_size)
            else:
                model = AdvancedRealEstateNN(self.input_size)
            
            model.load_state_dict(state_dict)
            model.to(device)
            model.eval()
            self.models[name] = model
        
        # ÿ™ÿ≠ŸÖŸäŸÑ scaler ÿ•ÿ∞ÿß ŸÉÿßŸÜ ŸÖŸàÿ¨ŸàÿØÿßŸã
        try:
            self.scaler = joblib.load('scaler.pkl')
        except:
            self.scaler = None
            print("‚ö†Ô∏è  Scaler not found. Make sure to scale input data manually.")
    
    def predict(self, new_data):
        """
        ÿßŸÑÿ™ŸÜÿ®ÿ§ ÿπŸÑŸâ ÿ®ŸäÿßŸÜÿßÿ™ ÿ¨ÿØŸäÿØÿ©
        
        Parameters:
        new_data: pandas DataFrame ÿ£Ÿà numpy array ÿ£Ÿà list
                  Ÿäÿ¨ÿ® ÿ£ŸÜ Ÿäÿ≠ÿ™ŸàŸä ÿπŸÑŸâ ŸÜŸÅÿ≥ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ÿßŸÑŸÖÿ≥ÿ™ÿÆÿØŸÖÿ© ŸÅŸä ÿßŸÑÿ™ÿØÿ±Ÿäÿ®
        """
        # ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ•ŸÑŸâ DataFrame
        if isinstance(new_data, pd.DataFrame):
            df = new_data.copy()
        elif isinstance(new_data, np.ndarray):
            df = pd.DataFrame(new_data, columns=self.feature_names[:new_data.shape[1]])
        elif isinstance(new_data, list):
            df = pd.DataFrame([new_data], columns=self.feature_names[:len(new_data)])
        else:
            raise ValueError("ŸÜŸàÿπ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ∫Ÿäÿ± ŸÖÿØÿπŸàŸÖ. ÿßÿ≥ÿ™ÿÆÿØŸÖ DataFrame, numpy array, ÿ£Ÿà list")
        
        # ÿßŸÑÿ™ÿ£ŸÉÿØ ŸÖŸÜ Ÿàÿ¨ŸàÿØ ÿ¨ŸÖŸäÿπ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ÿßŸÑŸÖÿ∑ŸÑŸàÿ®ÿ©
        missing_features = set(self.feature_names) - set(df.columns)
        if missing_features:
            raise ValueError(f"ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ÿßŸÑÿ™ÿßŸÑŸäÿ© ŸÖŸÅŸÇŸàÿØÿ©: {missing_features}")
        
        # ÿ•ÿπÿßÿØÿ© ÿ™ÿ±ÿ™Ÿäÿ® ÿßŸÑÿ£ÿπŸÖÿØÿ©
        df = df[self.feature_names]
        
        # ÿ™ÿ≠ŸàŸäŸÑ ÿ•ŸÑŸâ numpy
        data_np = df.values.astype(np.float32)
        
        # ÿ™ÿ∑ÿ®ŸäŸÇ scaling ÿ•ÿ∞ÿß ŸÉÿßŸÜ scaler ŸÖŸàÿ¨ŸàÿØÿßŸã
        if self.scaler:
            data_np = self.scaler.transform(data_np)
        
        # ÿ™ÿ≠ŸàŸäŸÑ ÿ•ŸÑŸâ tensor
        data_tensor = torch.FloatTensor(data_np).to(device)
        
        # ÿßŸÑÿ™ŸÜÿ®ÿ§ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿ¨ŸÖŸäÿπ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
        predictions = {}
        with torch.no_grad():
            for name, model in self.models.items():
                pred = model(data_tensor).cpu().numpy().flatten()
                predictions[name] = pred
        
        # ÿ≠ÿ≥ÿßÿ® ÿ™ŸÜÿ®ÿ§ÿßÿ™ Ensemble
        ensemble_pred = np.zeros_like(next(iter(predictions.values())))
        for name, pred in predictions.items():
            if name in self.weights:
                ensemble_pred += pred * self.weights[name]
        
        predictions['Ensemble'] = ensemble_pred
        
        return predictions

# ÿßÿÆÿ™ÿ®ÿßÿ± ÿØÿßŸÑÿ© ÿßŸÑÿ™ŸÜÿ®ÿ§
print("\n" + "="*60)
print("üîÆ TESTING PREDICTION FUNCTION")
print("="*60)

# ÿ•ŸÜÿ¥ÿßÿ° ÿπŸäŸÜÿ© ŸÑŸÑÿßÿÆÿ™ÿ®ÿßÿ±
sample_data = X_test.iloc[:3].copy()  # ÿ£ŸàŸÑ 3 ÿπŸäŸÜÿßÿ™ ŸÖŸÜ ŸÖÿ¨ŸÖŸàÿπÿ© ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±

try:
    predictor = RealEstatePredictor()
    predictions = predictor.predict(sample_data)
    
    print(f"\nüìù Sample Predictions (first 3 test samples):")
    print("-" * 50)
    
    for i in range(3):
        actual = y_test.iloc[i].values[0]
        ensemble_pred = predictions['Ensemble'][i]
        
        print(f"\nSample {i+1}:")
        print(f"  Actual value: {actual:.4f}")
        print(f"  Ensemble prediction: {ensemble_pred:.4f}")
        print(f"  Error: {ensemble_pred - actual:.4f}")
        print(f"  Error %: {abs((ensemble_pred - actual) / actual) * 100 if actual != 0 else 0:.2f}%")
    
    print("\n‚úÖ Prediction function is working correctly!")
    
except Exception as e:
    print(f"‚ö†Ô∏è  Error testing prediction function: {e}")

# ============================================
# üéâ 14. ÿπÿ±ÿ∂ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑŸÜŸáÿßÿ¶Ÿäÿ©
# ============================================
print("\n" + "="*60)
print("üéâ FINAL RESULTS SUMMARY")
print("="*60)

print(f"\nüìä OVERALL PERFORMANCE:")
print(f"  Best R¬≤ Score: {results_df['R2'].max():.4f}")
print(f"  Best RMSE: {results_df['RMSE'].min():.4f}")
print(f"  Best MAE: {results_df['MAE'].min():.4f}")

print(f"\nüìà IMPROVEMENT FROM BASELINE:")
original_r2 = 0.6277  # ŸÖŸÜ ŸÜÿ™ÿßÿ¶ÿ¨ŸÉ ÿßŸÑÿ≥ÿßÿ®ŸÇÿ©
best_r2 = results_df['R2'].max()
improvement = best_r2 - original_r2
print(f"  R¬≤ improvement: {improvement:+.4f} ({improvement/original_r2*100:+.1f}%)")

print(f"\nüíæ MODELS SAVED:")
print("  1. best_*.pth - Individual best models")
print("  2. real_estate_ensemble_final.pth - Complete ensemble model")
print("  3. model_performance_results.csv - Performance metrics")
print("  4. all_predictions.csv - All predictions")
print("  5. comprehensive_results.png - Visualizations")
print("  6. final_report.txt - Detailed report")

print(f"\nüöÄ NEXT STEPS:")
print("  1. Use RealEstatePredictor class for new predictions")
print("  2. Monitor model performance on new data")
print("  3. Consider hyperparameter tuning for further improvement")
print("  4. Implement model deployment if performance is satisfactory")

print("\n" + "="*60)
print("‚úÖ REAL ESTATE NEURAL NETWORK COMPLETED SUCCESSFULLY!")
print("="*60)
```
