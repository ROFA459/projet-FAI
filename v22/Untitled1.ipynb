```python
# ============================================
# üì¶ 1. ÿ™ÿ´ÿ®Ÿäÿ™ Ÿàÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÖŸÉÿ™ÿ®ÿßÿ™ ÿßŸÑŸÖÿ∑ŸÑŸàÿ®ÿ©
# ============================================
import subprocess
import sys

def install_packages():
    """ÿ™ÿ´ÿ®Ÿäÿ™ ÿßŸÑŸÖŸÉÿ™ÿ®ÿßÿ™ ÿßŸÑŸÖÿ∑ŸÑŸàÿ®ÿ© ÿ•ÿ∞ÿß ŸÑŸÖ ÿ™ŸÉŸÜ ŸÖÿ´ÿ®ÿ™ÿ©"""
    required_packages = [
        'torch', 'pandas', 'numpy', 'matplotlib',
        'seaborn', 'scikit-learn', 'scipy', 'joblib'
    ]
    
    for package in required_packages:
        try:
            if package == 'torch':
                __import__('torch')
            elif package == 'sklearn':
                __import__('sklearn')
            else:
                __import__(package)
            print(f"‚úÖ {package} already installed")
        except ImportError:
            print(f"üì¶ Installing {package}...")
            subprocess.check_call([sys.executable, "-m", "pip", "install", package, "--quiet"])

# ÿ™ÿ´ÿ®Ÿäÿ™ ÿßŸÑŸÖŸÉÿ™ÿ®ÿßÿ™
print("üîß Checking and installing required packages...")
install_packages()

# ============================================
# üìö 2. ÿßÿ≥ÿ™Ÿäÿ±ÿßÿØ ÿßŸÑŸÖŸÉÿ™ÿ®ÿßÿ™
# ============================================
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, Dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import warnings
import os
warnings.filterwarnings('ignore')

# ÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑÿπÿ±ÿ∂
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.4f}'.format)
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("‚úÖ All libraries imported successfully!")

# ============================================
# üîç 3. ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑŸÖÿπÿØÿ© ŸÖÿ≥ÿ®ŸÇÿßŸã
# ============================================
print("\n" + "="*60)
print("üìä LOADING PREPROCESSED DATA")
print("="*60)

try:
    # ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑŸÖŸÅŸÑÿ™ÿ±ÿ©
    X_train = pd.read_csv("train_filtered.csv")
    y_train = pd.read_csv("y_train.csv")
    X_val = pd.read_csv("val_filtered.csv")
    y_val = pd.read_csv("y_val.csv")
    X_test = pd.read_csv("test_filtered.csv")
    y_test = pd.read_csv("y_test.csv")
    
    print(f"‚úÖ Data loaded successfully:")
    print(f"  Training:   {X_train.shape}")
    print(f"  Validation: {X_val.shape}")
    print(f"  Test:       {X_test.shape}")
    
    print(f"\nüîç Features: {list(X_train.columns)}")
    print(f"üìä Target statistics (y_train):")
    print(f"  Mean: {y_train.values.mean():.2f}")
    print(f"  Std:  {y_train.values.std():.2f}")
    print(f"  Min:  {y_train.values.min():.2f}")
    print(f"  Max:  {y_train.values.max():.2f}")
    
except FileNotFoundError as e:
    print(f"‚ùå Error loading files: {e}")
    print("‚ö†Ô∏è  Make sure you have run the preprocessing code first")
    print("üìÅ Required files: train_filtered.csv, y_train.csv, val_filtered.csv, y_val.csv, test_filtered.csv, y_test.csv")
    import sys
    sys.exit(1)

# ============================================
# ‚öôÔ∏è 4. ÿ•ÿπÿØÿßÿØ ÿßŸÑÿ¨Ÿáÿßÿ≤ Ÿàÿ™ŸáŸäÿ¶ÿ© ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ============================================
# ÿ™ÿ≠ÿØŸäÿØ ÿßŸÑÿ¨Ÿáÿßÿ≤ (GPU ÿ•ÿ∞ÿß ŸÖÿ™ŸàŸÅÿ±)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\nüíª Using device: {device}")

# ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ•ŸÑŸâ tensors
def prepare_tensors(X_train, y_train, X_val, y_val, X_test, y_test):
    """ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ•ŸÑŸâ PyTorch tensors"""
    # ÿ™ÿ≠ŸàŸäŸÑ ÿ•ŸÑŸâ numpy arrays ÿ£ŸàŸÑÿßŸã
    X_train_np = X_train.values.astype(np.float32)
    y_train_np = y_train.values.astype(np.float32).reshape(-1, 1)
    X_val_np = X_val.values.astype(np.float32)
    y_val_np = y_val.values.astype(np.float32).reshape(-1, 1)
    X_test_np = X_test.values.astype(np.float32)
    y_test_np = y_test.values.astype(np.float32).reshape(-1, 1)
    
    # ÿ™ÿ≠ŸàŸäŸÑ ÿ•ŸÑŸâ PyTorch tensors
    X_train_tensor = torch.from_numpy(X_train_np).to(device)
    y_train_tensor = torch.from_numpy(y_train_np).to(device)
    X_val_tensor = torch.from_numpy(X_val_np).to(device)
    y_val_tensor = torch.from_numpy(y_val_np).to(device)
    X_test_tensor = torch.from_numpy(X_test_np).to(device)
    y_test_tensor = torch.from_numpy(y_test_np).to(device)
    
    return (X_train_tensor, y_train_tensor, X_val_tensor, y_val_tensor, 
            X_test_tensor, y_test_tensor)

# ÿ™ÿ≠ÿ∂Ÿäÿ± ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
tensors = prepare_tensors(X_train, y_train, X_val, y_val, X_test, y_test)
X_train_t, y_train_t, X_val_t, y_val_t, X_test_t, y_test_t = tensors

print(f"\n‚úÖ Data prepared successfully:")
print(f"  Training tensor shape:   {X_train_t.shape}")
print(f"  Validation tensor shape: {X_val_t.shape}")
print(f"  Test tensor shape:       {X_test_t.shape}")

# ============================================
# üß† 5. ÿ™ÿπÿ±ŸäŸÅ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨ ÿßŸÑÿπÿµÿ®Ÿäÿ©
# ============================================
print("\n" + "="*60)
print("üß† DEFINING NEURAL NETWORK MODELS")
print("="*60)

# ÿßÿ≥ÿ™Ÿäÿ±ÿßÿØ BatchNorm ÿ®ÿπÿØ ÿßŸÑÿ™ÿ£ŸÉÿØ ŸÖŸÜ ÿ£ŸÜŸÜÿß ŸÑÿ≥ŸÜÿß ŸÅŸä ÿßŸÑÿ≠ÿßŸÑÿ© ÿßŸÑÿ£ŸàŸÑŸâ
class SafeBatchNorm1d(nn.BatchNorm1d):
    """BatchNorm ÿ¢ŸÖŸÜ Ÿäÿ™ÿπÿßŸÖŸÑ ŸÖÿπ batch_size=1"""
    def forward(self, input):
        if input.size(0) > 1:
            return super().forward(input)
        else:
            # ÿ•ÿ∞ÿß ŸÉÿßŸÜ batch_size=1ÿå ŸÜÿ™ÿ¨ÿßŸàÿ≤ BatchNorm
            return input

class SimpleRealEstateNN(nn.Module):
    """ŸÜŸÖŸàÿ∞ÿ¨ ÿ®ÿ≥Ÿäÿ∑ ŸàŸÅÿπÿßŸÑ"""
    def __init__(self, input_size):
        super(SimpleRealEstateNN, self).__init__()
        
        self.network = nn.Sequential(
            nn.Linear(input_size, 32),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.2),
            
            nn.Linear(32, 16),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.2),
            
            nn.Linear(16, 8),
            nn.LeakyReLU(0.1),
            
            nn.Linear(8, 1)
        )
    
    def forward(self, x):
        return self.network(x)

class MediumRealEstateNN(nn.Module):
    """ŸÜŸÖŸàÿ∞ÿ¨ ŸÖÿ™Ÿàÿ≥ÿ∑ ÿßŸÑÿ≠ÿ¨ŸÖ"""
    def __init__(self, input_size):
        super(MediumRealEstateNN, self).__init__()
        
        self.network = nn.Sequential(
            nn.Linear(input_size, 64),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.3),
            
            nn.Linear(64, 32),
            SafeBatchNorm1d(32) if input_size > 1 else nn.Identity(),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.3),
            
            nn.Linear(32, 16),
            nn.LeakyReLU(0.1),
            
            nn.Linear(16, 1)
        )
    
    def forward(self, x):
        return self.network(x)

class AdvancedRealEstateNN(nn.Module):
    """ŸÜŸÖŸàÿ∞ÿ¨ ŸÖÿ™ŸÇÿØŸÖ"""
    def __init__(self, input_size):
        super(AdvancedRealEstateNN, self).__init__()
        
        self.network = nn.Sequential(
            nn.Linear(input_size, 128),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.4),
            
            nn.Linear(128, 64),
            SafeBatchNorm1d(64) if input_size > 1 else nn.Identity(),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.3),
            
            nn.Linear(64, 32),
            SafeBatchNorm1d(32) if input_size > 1 else nn.Identity(),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.2),
            
            nn.Linear(32, 16),
            nn.LeakyReLU(0.1),
            
            nn.Linear(16, 1)
        )
    
    def forward(self, x):
        return self.network(x)

# ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
input_size = X_train.shape[1]
models = {
    'Simple': SimpleRealEstateNN(input_size),
    'Medium': MediumRealEstateNN(input_size),
    'Advanced': AdvancedRealEstateNN(input_size)
}

for name, model in models.items():
    model.to(device)
    print(f"‚úÖ {name} Model created: {sum(p.numel() for p in model.parameters()):,} parameters")

# ============================================
# üîÑ 6. ÿ•ŸÜÿ¥ÿßÿ° DataLoaders
# ============================================
print("\n" + "="*60)
print("üîÑ CREATING DATALOADERS")
print("="*60)

class RealEstateDataset(Dataset):
    """Dataset ŸÑŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿπŸÇÿßÿ±Ÿäÿ©"""
    def __init__(self, features, targets):
        self.features = features
        self.targets = targets
    
    def __len__(self):
        return len(self.features)
    
    def __getitem__(self, idx):
        return self.features[idx], self.targets[idx]

# ÿ•ŸÜÿ¥ÿßÿ° datasets
train_dataset = RealEstateDataset(X_train_t, y_train_t)
val_dataset = RealEstateDataset(X_val_t, y_val_t)
test_dataset = RealEstateDataset(X_test_t, y_test_t)

# ÿ≠ÿ≥ÿßÿ® batch size ÿØŸäŸÜÿßŸÖŸäŸÉŸäÿßŸã
def calculate_batch_size(dataset_size):
    """ÿ≠ÿ≥ÿßÿ® batch size ÿØŸäŸÜÿßŸÖŸäŸÉŸäÿßŸã"""
    if dataset_size >= 128:
        return 64
    elif dataset_size >= 64:
        return 32
    elif dataset_size >= 32:
        return 16
    elif dataset_size >= 16:
        return 8
    else:
        return max(4, dataset_size // 2)

batch_size = calculate_batch_size(len(X_train_t))
print(f"üìä Using batch size: {batch_size}")

# ÿ•ŸÜÿ¥ÿßÿ° DataLoaders
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print(f"‚úÖ DataLoaders created:")
print(f"  Training batches: {len(train_loader)}")
print(f"  Validation batches: {len(val_loader)}")
print(f"  Test batches: {len(test_loader)}")

# ============================================
# üèãÔ∏è 7. ŸÅÿ¶ÿ© ÿßŸÑÿ™ÿØÿ±Ÿäÿ®
# ============================================
class RealEstateTrainer:
    """ŸÅÿ¶ÿ© ŸÖÿ™ŸÇÿØŸÖÿ© ŸÑŸÑÿ™ÿØÿ±Ÿäÿ®"""
    def __init__(self, model, model_name="Model"):
        self.model = model.to(device)
        self.model_name = model_name
        self.train_losses = []
        self.val_losses = []
        self.learning_rates = []
        self.best_val_loss = float('inf')
        self.best_model_state = None
    
    def train(self, train_loader, val_loader, epochs=200, lr=0.001, weight_decay=0.0001):
        """ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨"""
        print(f"\nüöÄ Training {self.model_name} for {epochs} epochs...")
        print(f"  Learning rate: {lr}, Weight decay: {weight_decay}")
        
        criterion = nn.MSELoss()
        optimizer = optim.Adam(self.model.parameters(), lr=lr, weight_decay=weight_decay)
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='min', patience=10, factor=0.5, verbose=True
        )
        
        for epoch in range(epochs):
            # Training phase
            self.model.train()
            train_loss = 0.0
            
            for batch_X, batch_y in train_loader:
                optimizer.zero_grad()
                
                # Forward pass
                output = self.model(batch_X)
                loss = criterion(output, batch_y)
                
                # Backward pass
                loss.backward()
                
                # Gradient clipping
                torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
                
                optimizer.step()
                train_loss += loss.item()
            
            avg_train_loss = train_loss / len(train_loader)
            self.train_losses.append(avg_train_loss)
            
            # Validation phase
            self.model.eval()
            val_loss = 0.0
            
            with torch.no_grad():
                for batch_X, batch_y in val_loader:
                    output = self.model(batch_X)
                    val_loss += criterion(output, batch_y).item()
            
            avg_val_loss = val_loss / len(val_loader)
            self.val_losses.append(avg_val_loss)
            
            # Update scheduler
            scheduler.step(avg_val_loss)
            
            # Save learning rate
            self.learning_rates.append(optimizer.param_groups[0]['lr'])
            
            # Save best model
            if avg_val_loss < self.best_val_loss:
                self.best_val_loss = avg_val_loss
                self.best_model_state = self.model.state_dict().copy()
                torch.save(self.model.state_dict(), f'best_{self.model_name}.pth')
            
            # Print progress
            if (epoch + 1) % 25 == 0 or epoch == 0 or epoch == epochs - 1:
                print(f"  Epoch [{epoch+1:03d}/{epochs}] | "
                      f"Train Loss: {avg_train_loss:.6f} | "
                      f"Val Loss: {avg_val_loss:.6f} | "
                      f"LR: {self.learning_rates[-1]:.6f}")
        
        # Load best model
        self.model.load_state_dict(self.best_model_state)
        
        print(f"‚úÖ {self.model_name} training completed!")
        print(f"üìâ Best validation loss: {self.best_val_loss:.6f}")
        
        return self.model
    
    def plot_history(self):
        """ÿ±ÿ≥ŸÖ ŸÖŸÜÿ≠ŸÜŸâ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®"""
        if len(self.train_losses) == 0:
            print("‚ö†Ô∏è  No training history to plot")
            return
            
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
        
        # ÿ±ÿ≥ŸÖ ÿßŸÑÿÆÿ≥ÿßÿ¶ÿ±
        ax1.plot(self.train_losses, label='Training Loss', linewidth=2)
        ax1.plot(self.val_losses, label='Validation Loss', linewidth=2)
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss (MSE)')
        ax1.set_title(f'{self.model_name} - Training History')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # ÿ±ÿ≥ŸÖ ŸÖÿπÿØŸÑ ÿßŸÑÿ™ÿπŸÑŸÖ
        ax2.plot(self.learning_rates, color='green', linewidth=2)
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Learning Rate')
        ax2.set_title('Learning Rate Schedule')
        ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(f'training_{self.model_name}.png', dpi=120)
        plt.show()

# ============================================
# ü§ñ 8. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
# ============================================
print("\n" + "="*60)
print("ü§ñ TRAINING MODELS")
print("="*60)

# ÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®
training_configs = {
    'Simple': {'epochs': 150, 'lr': 0.0015, 'weight_decay': 0.0001},
    'Medium': {'epochs': 200, 'lr': 0.001, 'weight_decay': 0.0001},
    'Advanced': {'epochs': 250, 'lr': 0.0008, 'weight_decay': 0.0001}
}

# ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
trained_models = {}
for model_name, model in models.items():
    print(f"\n{'='*40}")
    print(f"üî• Training {model_name} model")
    print('='*40)
    
    try:
        trainer = RealEstateTrainer(model, model_name)
        
        trained_model = trainer.train(
            train_loader=train_loader,
            val_loader=val_loader,
            epochs=training_configs[model_name]['epochs'],
            lr=training_configs[model_name]['lr'],
            weight_decay=training_configs[model_name]['weight_decay']
        )
        
        trainer.plot_history()
        trained_models[model_name] = {
            'model': trained_model,
            'trainer': trainer
        }
        
        print(f"‚úÖ {model_name} trained successfully!")
        
    except Exception as e:
        print(f"‚ùå Error training {model_name}: {e}")
        import traceback
        traceback.print_exc()

# ============================================
# üìä 9. ÿ™ŸÇŸäŸäŸÖ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
# ============================================
print("\n" + "="*60)
print("üìä EVALUATING MODELS")
print("="*60)

def evaluate_model(model, test_loader):
    """ÿ™ŸÇŸäŸäŸÖ ÿ¥ÿßŸÖŸÑ ŸÑŸÑŸÜŸÖŸàÿ∞ÿ¨"""
    model.eval()
    all_predictions = []
    all_targets = []
    
    with torch.no_grad():
        for data, target in test_loader:
            output = model(data)
            all_predictions.extend(output.cpu().numpy())
            all_targets.extend(target.cpu().numpy())
    
    all_predictions = np.array(all_predictions).flatten()
    all_targets = np.array(all_targets).flatten()
    
    # ÿ≠ÿ≥ÿßÿ® ÿßŸÑŸÖŸÇÿßŸäŸäÿ≥
    metrics = {
        'MSE': mean_squared_error(all_targets, all_predictions),
        'RMSE': np.sqrt(mean_squared_error(all_targets, all_predictions)),
        'MAE': mean_absolute_error(all_targets, all_predictions),
        'R2': r2_score(all_targets, all_predictions),
        'MAPE': np.mean(np.abs((all_targets - all_predictions) / np.clip(np.abs(all_targets), 1e-8, None))) * 100
    }
    
    return metrics, all_predictions, all_targets

# ÿ™ŸÇŸäŸäŸÖ ÿ¨ŸÖŸäÿπ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
results = {}
all_model_predictions = {}
test_targets = None

if len(trained_models) > 0:
    for name, model_info in trained_models.items():
        print(f"\nüìà Evaluating {name} model...")
        
        metrics, predictions, targets = evaluate_model(model_info['model'], test_loader)
        results[name] = metrics
        all_model_predictions[name] = predictions
        
        # ÿ≠ŸÅÿ∏ targets ŸÑÿ£ŸàŸÑ ŸÖÿ±ÿ©
        if test_targets is None:
            test_targets = targets
        
        print(f"  R¬≤ Score:   {metrics['R2']:.4f}")
        print(f"  RMSE:       {metrics['RMSE']:.4f}")
        print(f"  MAE:        {metrics['MAE']:.4f}")
        print(f"  MAPE:       {metrics['MAPE']:.2f}%")
else:
    print("‚ùå No models were trained successfully!")
    print("‚ö†Ô∏è  Exiting...")
    import sys
    sys.exit(1)

# ============================================
# ü§ù 10. Ensemble Model
# ============================================
print("\n" + "="*60)
print("üèÜ ENSEMBLE MODEL")
print("="*60)

# ÿ•ŸÜÿ¥ÿßÿ° Ensemble ŸÖŸÜ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
if len(all_model_predictions) > 0 and test_targets is not None:
    # ÿ¨ŸÖÿπ ÿ™ŸÜÿ®ÿ§ÿßÿ™ ÿ¨ŸÖŸäÿπ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
    ensemble_predictions = np.zeros_like(list(all_model_predictions.values())[0])
    
    # ÿ£Ÿàÿ≤ÿßŸÜ ÿ™ÿπÿ™ŸÖÿØ ÿπŸÑŸâ ÿ£ÿØÿßÿ° R¬≤
    weights = {}
    total_r2 = sum(results[name]['R2'] for name in results)
    
    for name in results:
        weights[name] = results[name]['R2'] / total_r2
        ensemble_predictions += all_model_predictions[name] * weights[name]
    
    # ÿ™ŸÇŸäŸäŸÖ Ensemble
    ensemble_metrics = {
        'MSE': mean_squared_error(test_targets, ensemble_predictions),
        'RMSE': np.sqrt(mean_squared_error(test_targets, ensemble_predictions)),
        'MAE': mean_absolute_error(test_targets, ensemble_predictions),
        'R2': r2_score(test_targets, ensemble_predictions),
        'MAPE': np.mean(np.abs((test_targets - ensemble_predictions) / np.clip(np.abs(test_targets), 1e-8, None))) * 100
    }
    
    results['Ensemble'] = ensemble_metrics
    all_model_predictions['Ensemble'] = ensemble_predictions
    
    print(f"\nüéØ Ensemble Model Results:")
    print(f"  R¬≤ Score:   {ensemble_metrics['R2']:.4f}")
    print(f"  RMSE:       {ensemble_metrics['RMSE']:.4f}")
    print(f"  MAE:        {ensemble_metrics['MAE']:.4f}")
    print(f"  MAPE:       {ensemble_metrics['MAPE']:.2f}%")
    
    print(f"\nüìä Model Weights:")
    for name, weight in weights.items():
        print(f"  {name}: {weight:.3f} (R¬≤: {results[name]['R2']:.4f})")

# ============================================
# üìà 11. ÿπÿ±ÿ∂ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨ ÿ®ŸäÿßŸÜŸäÿßŸã
# ============================================
print("\n" + "="*60)
print("üìà VISUALIZING RESULTS")
print("="*60)

def create_visualizations(results, predictions_dict, targets):
    """ÿ•ŸÜÿ¥ÿßÿ° ÿ±ÿ≥ŸàŸÖ ÿ®ŸäÿßŸÜŸäÿ© ÿ¥ÿßŸÖŸÑÿ©"""
    fig = plt.figure(figsize=(18, 12))
    
    # 1. ŸÖŸÇÿßÿ±ŸÜÿ© ÿ£ÿØÿßÿ° ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
    ax1 = plt.subplot(2, 3, 1)
    model_names = list(results.keys())
    r2_scores = [results[name]['R2'] for name in model_names]
    rmse_scores = [results[name]['RMSE'] for name in model_names]
    
    x = np.arange(len(model_names))
    width = 0.35
    
    bars1 = ax1.bar(x - width/2, r2_scores, width, label='R¬≤ Score', color='skyblue')
    bars2 = ax1.bar(x + width/2, rmse_scores, width, label='RMSE', color='lightcoral')
    
    ax1.set_xlabel('Models')
    ax1.set_title('Model Performance Comparison')
    ax1.set_xticks(x)
    ax1.set_xticklabels(model_names, rotation=45, ha='right')
    ax1.legend()
    ax1.grid(True, alpha=0.3, axis='y')
    
    # ÿ•ÿ∂ÿßŸÅÿ© ÿßŸÑŸÇŸäŸÖ ÿπŸÑŸâ ÿßŸÑÿ£ÿπŸÖÿØÿ©
    for bar, score in zip(bars1, r2_scores):
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height,
                f'{score:.3f}', ha='center', va='bottom', fontsize=9)
    
    # 2. ÿßŸÑÿ™ŸÜÿ®ÿ§ÿßÿ™ ŸÖŸÇÿßÿ®ŸÑ ÿßŸÑŸÇŸäŸÖ ÿßŸÑÿ≠ŸÇŸäŸÇŸäÿ© (Ensemble)
    ax2 = plt.subplot(2, 3, 2)
    if 'Ensemble' in predictions_dict:
        y_pred = predictions_dict['Ensemble']
        ax2.scatter(targets, y_pred, alpha=0.6, s=30)
        
        min_val = min(targets.min(), y_pred.min())
        max_val = max(targets.max(), y_pred.max())
        ax2.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2)
        
        ax2.set_xlabel('Actual Values')
        ax2.set_ylabel('Predicted Values')
        ax2.set_title(f'Ensemble Predictions\nR¬≤ = {results["Ensemble"]["R2"]:.4f}')
        ax2.grid(True, alpha=0.3)
    
    # 3. ÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑÿ£ÿÆÿ∑ÿßÿ°
    ax3 = plt.subplot(2, 3, 3)
    if 'Ensemble' in predictions_dict:
        errors = predictions_dict['Ensemble'] - targets
        ax3.hist(errors, bins=30, edgecolor='black', alpha=0.7, density=True)
        ax3.axvline(x=0, color='r', linestyle='--', linewidth=2)
        
        # ÿ•ÿ∂ÿßŸÅÿ© ŸÖŸÜÿ≠ŸÜŸâ ÿßŸÑÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑÿ∑ÿ®ŸäÿπŸä
        mu, sigma = errors.mean(), errors.std()
        x_normal = np.linspace(errors.min(), errors.max(), 100)
        y_normal = (1/(sigma * np.sqrt(2*np.pi))) * np.exp(-0.5*((x_normal-mu)/sigma)**2)
        ax3.plot(x_normal, y_normal, 'r-', linewidth=2)
        
        ax3.set_xlabel('Prediction Error')
        ax3.set_ylabel('Density')
        ax3.set_title(f'Error Distribution\nMean: {mu:.3f}, Std: {sigma:.3f}')
        ax3.grid(True, alpha=0.3)
    
    # 4. Residual Plot
    ax4 = plt.subplot(2, 3, 4)
    if 'Ensemble' in predictions_dict:
        residuals = predictions_dict['Ensemble'] - targets
        ax4.scatter(predictions_dict['Ensemble'], residuals, alpha=0.6, s=30)
        ax4.axhline(y=0, color='r', linestyle='--', linewidth=2)
        ax4.set_xlabel('Predicted Values')
        ax4.set_ylabel('Residuals')
        ax4.set_title('Residual Plot')
        ax4.grid(True, alpha=0.3)
    
    # 5. ŸÖŸÇÿßÿ±ŸÜÿ© ÿßŸÑÿ™ŸÜÿ®ÿ§ÿßÿ™
    ax5 = plt.subplot(2, 3, 5)
    sample_size = min(15, len(targets))
    indices = np.arange(sample_size)
    
    # ÿ±ÿ≥ŸÖ ÿ£ŸàŸÑ 15 ÿπŸäŸÜÿ©
    for name, preds in predictions_dict.items():
        if name != 'Ensemble':
            ax5.plot(indices, preds[:sample_size], 'o-', label=name, alpha=0.7, linewidth=1.5)
    
    ax5.plot(indices, targets[:sample_size], 'ks-', label='Actual', linewidth=2, markersize=6)
    ax5.set_xlabel('Sample Index')
    ax5.set_ylabel('Value')
    ax5.set_title(f'Model Predictions (First {sample_size} Samples)')
    ax5.legend(loc='upper right', fontsize=8)
    ax5.grid(True, alpha=0.3)
    
    # 6. Box plot ŸÑŸÑÿ£ÿÆÿ∑ÿßÿ°
    ax6 = plt.subplot(2, 3, 6)
    error_data = []
    labels = []
    
    for name, preds in predictions_dict.items():
        if name != 'Ensemble':  # ŸÜÿ≥ÿ™ÿ´ŸÜŸä Ensemble ŸÑŸÑŸÖŸÇÿßÿ±ŸÜÿ©
            error_data.append(preds - targets)
            labels.append(name)
    
    if error_data:
        bp = ax6.boxplot(error_data, labels=labels, patch_artist=True)
        
        # ÿ™ŸÑŸàŸäŸÜ ÿßŸÑÿµŸÜÿßÿØŸäŸÇ
        colors = ['lightblue', 'lightgreen', 'lightcoral']
        for patch, color in zip(bp['boxes'], colors[:len(error_data)]):
            patch.set_facecolor(color)
        
        ax6.axhline(y=0, color='r', linestyle='--', linewidth=2)
        ax6.set_ylabel('Prediction Error')
        ax6.set_title('Error Distribution by Model')
        ax6.tick_params(axis='x', rotation=45)
        ax6.grid(True, alpha=0.3, axis='y')
    
    plt.suptitle('Real Estate Neural Network - Comprehensive Results Analysis', 
                 fontsize=16, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.savefig('final_results_comprehensive.png', dpi=150, bbox_inches='tight')
    plt.show()

# ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑÿ±ÿ≥ŸàŸÖ ÿßŸÑÿ®ŸäÿßŸÜŸäÿ©
if test_targets is not None and len(all_model_predictions) > 0:
    create_visualizations(results, all_model_predictions, test_targets)
else:
    print("‚ö†Ô∏è  Not enough data for visualizations")

# ============================================
# üíæ 12. ÿ≠ŸÅÿ∏ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨ ŸàÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
# ============================================
print("\n" + "="*60)
print("üíæ SAVING RESULTS AND MODELS")
print("="*60)

# 1. ÿ≠ŸÅÿ∏ ŸÖŸÇÿßŸäŸäÿ≥ ÿßŸÑÿ£ÿØÿßÿ°
results_df = pd.DataFrame(results).T
results_df.to_csv('model_performance_metrics.csv')
print("‚úÖ Model performance metrics saved to 'model_performance_metrics.csv'")

# 2. ÿ≠ŸÅÿ∏ ÿßŸÑÿ™ŸÜÿ®ÿ§ÿßÿ™
if test_targets is not None and len(all_model_predictions) > 0:
    predictions_df = pd.DataFrame({
        'Actual': test_targets,
        'Ensemble_Prediction': all_model_predictions.get('Ensemble', np.zeros_like(test_targets))
    })
    
    # ÿ•ÿ∂ÿßŸÅÿ© ÿ™ŸÜÿ®ÿ§ÿßÿ™ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨ ÿßŸÑÿ£ÿÆÿ±Ÿâ
    for name, preds in all_model_predictions.items():
        if name != 'Ensemble':
            predictions_df[f'{name}_Prediction'] = preds
    
    predictions_df.to_csv('all_predictions.csv', index=False)
    print("‚úÖ All predictions saved to 'all_predictions.csv'")

# 3. ÿ≠ŸÅÿ∏ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿßŸÑŸÜŸáÿßÿ¶Ÿä
if 'Ensemble' in results and len(trained_models) > 0:
    # ÿ≠ŸÅÿ∏ ÿ≠ÿßŸÑÿ© Ensemble
    ensemble_state = {
        'model_states': {},
        'weights': weights if 'weights' in locals() else {},
        'input_size': input_size,
        'feature_names': list(X_train.columns),
        'results': results,
        'ensemble_metrics': results.get('Ensemble', {}),
        'timestamp': pd.Timestamp.now().isoformat()
    }
    
    # ÿ≠ŸÅÿ∏ ÿ≠ÿßŸÑÿßÿ™ ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨ ÿßŸÑŸÅÿ±ÿØŸäÿ©
    for name, model_info in trained_models.items():
        ensemble_state['model_states'][name] = model_info['model'].state_dict()
    
    torch.save(ensemble_state, 'real_estate_final_model.pth')
    print("‚úÖ Final ensemble model saved to 'real_estate_final_model.pth'")

# 4. ÿ•ŸÜÿ¥ÿßÿ° ÿ™ŸÇÿ±Ÿäÿ± ŸÜŸáÿßÿ¶Ÿä
print("\n" + "="*60)
print("üìÑ GENERATING FINAL REPORT")
print("="*60)

report_content = f"""
{'='*70}
REAL ESTATE NEURAL NETWORK - FINAL REPORT
{'='*70}

üìÖ Generated on: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}
üíª Device used: {device}
üìä Dataset: Real Estate Prediction

üìà DATA SUMMARY:
‚Ä¢ Training samples: {len(X_train)}
‚Ä¢ Validation samples: {len(X_val)}
‚Ä¢ Test samples: {len(X_test)}
‚Ä¢ Number of features: {input_size}
‚Ä¢ Features used: {list(X_train.columns)}

üèÜ PERFORMANCE RESULTS:
"""

for model_name, metrics in results.items():
    report_content += f"\n{model_name}:\n"
    report_content += f"  ‚Ä¢ R¬≤ Score: {metrics['R2']:.4f}\n"
    report_content += f"  ‚Ä¢ RMSE: {metrics['RMSE']:.4f}\n"
    report_content += f"  ‚Ä¢ MAE: {metrics['MAE']:.4f}\n"
    if 'MAPE' in metrics:
        report_content += f"  ‚Ä¢ MAPE: {metrics['MAPE']:.2f}%\n"

report_content += f"""
üìà IMPROVEMENT ANALYSIS:
‚Ä¢ Best model: {max(results.items(), key=lambda x: x[1]['R2'])[0]}
‚Ä¢ Best R¬≤: {max(m['R2'] for m in results.values()):.4f}
‚Ä¢ Ensemble R¬≤: {results.get('Ensemble', {}).get('R2', 0):.4f}

üíæ FILES GENERATED:
1. training_*.png - Training history plots
2. final_results_comprehensive.png - Comprehensive visualizations
3. model_performance_metrics.csv - Performance metrics
4. all_predictions.csv - All model predictions
5. real_estate_final_model.pth - Final trained model
6. best_*.pth - Best individual models

üöÄ RECOMMENDATIONS:
"""

best_r2 = max(m['R2'] for m in results.values())
if best_r2 > 0.8:
    report_content += "‚úÖ Excellent model performance! Ready for production use.\n"
elif best_r2 > 0.7:
    report_content += "‚ö†Ô∏è  Good performance. Consider feature engineering for further improvement.\n"
elif best_r2 > 0.6:
    report_content += "‚ö†Ô∏è  Fair performance. Try collecting more data or using different architectures.\n"
else:
    report_content += "‚ùå Needs improvement. Review data quality and preprocessing steps.\n"

report_content += f"\n{'='*70}"

# ÿ≠ŸÅÿ∏ ÿßŸÑÿ™ŸÇÿ±Ÿäÿ±
with open('final_report.txt', 'w', encoding='utf-8') as f:
    f.write(report_content)

print("‚úÖ Final report saved to 'final_report.txt'")

# ============================================
# üéâ 13. ÿπÿ±ÿ∂ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑŸÜŸáÿßÿ¶Ÿäÿ©
# ============================================
print("\n" + "="*60)
print("üéâ FINAL SUMMARY")
print("="*60)

print(f"\nüìä DATASET INFORMATION:")
print(f"  Features: {input_size}")
print(f"  Training samples: {len(X_train)}")
print(f"  Test samples: {len(X_test)}")

if results:
    print(f"\nüèÜ BEST MODEL PERFORMANCE:")
    best_model_name = max(results.items(), key=lambda x: x[1]['R2'])[0]
    best_metrics = results[best_model_name]
    print(f"  Model: {best_model_name}")
    print(f"  R¬≤ Score: {best_metrics['R2']:.4f}")
    print(f"  RMSE: {best_metrics['RMSE']:.4f}")
    print(f"  MAE: {best_metrics['MAE']:.4f}")

if 'Ensemble' in results:
    print(f"\nü§ù ENSEMBLE PERFORMANCE:")
    ensemble_metrics = results['Ensemble']
    print(f"  R¬≤ Score: {ensemble_metrics['R2']:.4f}")
    print(f"  RMSE: {ensemble_metrics['RMSE']:.4f}")
    print(f"  MAE: {ensemble_metrics['MAE']:.4f}")

print(f"\nüíæ FILES CREATED:")
print("  1. training_*.png - Training histories")
print("  2. final_results_comprehensive.png - Visualizations")
print("  3. model_performance_metrics.csv - Performance metrics")
print("  4. all_predictions.csv - All predictions")
if os.path.exists('real_estate_final_model.pth'):
    print("  5. real_estate_final_model.pth - Final model")
print("  6. final_report.txt - Detailed report")
print("  7. best_*.pth - Best individual models")

print(f"\nüöÄ NEXT STEPS:")
print("  1. Review the performance metrics in model_performance_metrics.csv")
print("  2. Check the visualizations in final_results_comprehensive.png")
print("  3. Use the trained model for new predictions")
print("  4. Monitor model performance on new data")

print("\n" + "="*60)
print("‚úÖ REAL ESTATE NEURAL NETWORK COMPLETED SUCCESSFULLY!")
print("="*60)
```
